```python
# -*- coding: utf-8 -*
import pandas as pd
import io
from unidecode import unidecode
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import json
```

# 7e continent

```python
new_articles = []
vectorizer = TfidfVectorizer()
# Get scraped articles
articles = json.loads(io.open('../data/articles/clean_art_7econ.json', 'r', encoding='utf-8-sig').read())

for i in articles:
    content = articles[str(i)]['content']
    X = vectorizer.fit_transform(word_tokenize(unidecode(content)))
    print(X.toarray()[0])
                          
    new_articles.append({
        "id_article":articles[str(i)]['id_article'],
        "link":articles[str(i)]['link'],
        "title":articles[str(i)]['title'],
        "content":articles[str(i)]['content'],
        "authors":articles[str(i)]['authors'],
        "newspaper":articles[str(i)]['newspaper'],
        "publication_date":articles[str(i)]['publication_date'],
        "words":'stub'
    })
```
